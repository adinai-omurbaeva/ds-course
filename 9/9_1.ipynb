Какой вариант оценки доверительных интервалов нужно использовать в данной задаче и почему? Какая из ошибок более критична для данной задачи — I или II рода?

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
df = pd.read_csv('train.csv')
df
```

Функция для вычисления доверительного интервала

```python
def compute_confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean
    interval = sem * stats.t.ppf((1 + confidence) / 2., len(data)-1)
    return mean, mean - interval, mean + interval
```

```python
# Выживаемость всех пассажиров
mean_all, lower_all, upper_all = compute_confidence_interval(df['Survived'])
print(f"Общая выживаемость: Среднее = {mean_all:.4f}, 95% ДИ = [{lower_all:.4f}, {upper_all:.4f}]")
```

```python
# Выживаемость женщин
mean_female, lower_female, upper_female = compute_confidence_interval(df[df['Sex'] == 'female']['Survived'])
print(f"Выживаемость женщин: Среднее = {mean_female:.4f}, 95% ДИ = [{lower_female:.4f}, {upper_female:.4f}]")
```

```python
# Выживаемость пассажиров первого класса
mean_first_class, lower_first_class, upper_first_class = compute_confidence_interval(df[df['Pclass'] == 1]['Survived'])
print(f"Выживаемость пассажиров первого класса: Среднее = {mean_first_class:.4f}, 95% ДИ = [{lower_first_class:.4f}, {upper_first_class:.4f}]")
```

Ошибка I рода (ложноположительный результат) более критична в данной задаче. Ошибка I рода означает, что мы ошибочно отвергаем нулевую гипотезу, 
когда она верна. В контексте выживаемости пассажиров, это может привести к неверным выводам о значимости определенных факторов, влияющих на 
выживаемость, что, в свою очередь, может привести к неправильным решениям по улучшению безопасности. Например, если мы ошибочно решим, что определенный 
фактор (например, класс каюты) значительно влияет на выживаемость, мы можем сосредоточиться на этом факторе и пренебречь другими, возможно более 
важными мерами безопасности.
