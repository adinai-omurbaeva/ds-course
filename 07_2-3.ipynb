Используя градиентный спуск найти минимум функции с точностью e = 10**-8ipy.



```python
import sympy as sym
import numpy as np
```

```python
def f(x, y):
     return (x - 3) ** 2 + (y + 5) ** 2

def grad_f(x, y):
    grad_x = 2 * x - 6 
    grad_y = 2 * y + 10
    
    return np.array([grad_x, grad_y])

eps = 10e-8
learning_rate = 0.1
max_steps = 1000

i = 0
x0 = np.array([5, 2])
x_old = x0
while np.linalg.norm( grad_f(x_old[0], x_old[1])) > eps:
    x_new = x_old - learning_rate *  grad_f(x_old[0], x_old[1])
    x_old = x_new
    i += 1
    if i == max_steps: break
    
print(x_new, f(x_new[0], x_new[1]),i)
```

Проверить себя при помощи функции minimize в scipy.

```python
from scipy.optimize import minimize
import numpy as np

def f(x):
    return (x[0] - 3) ** 2 + (x[1] + 5) ** 2

x0 = np.array([5, 2])
result = minimize(f, x0, method='BFGS', tol=1e-8)
print(result.x, result.fun, result.nit)
```

Для функции из прошлого задания построить график линий уровня.

```python
import matplotlib.pyplot as plt

def draw_graf():
    x = np.arange(-10000, 10000, 10)
    y = np.arange(-10000, 10000, 10)
    xgrid, ygrid = np.meshgrid(x, y)

    z = xgrid * xgrid  - 6 * xgrid + ygrid * ygrid + 10 * ygrid + 34
   
    return xgrid, ygrid, z

x, y, z = draw_graf()

plt.contourf(x, y, z, levels=20)
plt.xlabel('x')
plt.ylabel('y')
plt.title('График линий для f(x, y)')
plt.show()
```
